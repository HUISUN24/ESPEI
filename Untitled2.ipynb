```python
# Only needed in a Jupyter Notebook
%matplotlib inline
# Optional plot styling
import matplotlib
matplotlib.style.use('bmh')
import matplotlib.pyplot as plt
from pycalphad import equilibrium
from pycalphad import Database, Model
import pycalphad.variables as v
import numpy as np
```

```python
db = Database('mcmc_Y_try_adj3_fin.TDB')
my_phases = list(db.phases.keys())
eq = equilibrium(db, ['NB', 'NI', 'VA'], my_phases, {v.X('NB'): (0,1,0.01), v.T: 298.15, v.P: 101325},
                 output='HM')
print(eq)
```

```python
plt.gca().set_title('NB-NI: Degree of bcc ordering vs T [X(AL)=0.25]')
plt.gca().set_xlabel('x(NB)')
plt.gca().set_ylabel('HM')
plt.gca().set_xlim((0,1))
# Generate a list of all indices where B2 is stable
print(eq['X_NB'].values,eq['HM'].values)
# phase_indices[2] refers to all temperature indices
# We know this because pycalphad always returns indices in order like P, T, X's
plt.plot(eq['X_NB'].values, eq['HM'].values[0][0][0])
np.savetxt('result.txt',[eq['X_NB'].values, eq['HM'].values[0][0][0]])
plt.show()
```
